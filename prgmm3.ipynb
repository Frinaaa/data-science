```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([3, 6])
y = np.array([3, 8])

plt.subplot(2, 3, 1)
plt.plot(x,y,'o',linewidth=0.5)
plt.xlabel("x label",color='#880806')
plt.ylabel("y label",color='#880806')

font1 = {'family':'serif','size':15,'color':'#880806'}
plt.title("Graph1",fontdict = font1)
 
x = np.array([1, 2, 3, 4, 5])
y1 = np.array([2, 4, 2, 4, 2])
y2=np.array([3,1,4,1,3])

plt.subplot(2, 3, 2)
plt.plot(x,y1,color='orange',linewidth=2.5,marker="*")
plt.plot(x,y2,color='blue',linestyle="--")
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 3)
plt.plot(x,y,marker='o',mfc='yellow',mec='black',color='blue')

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.title("Graph 4",color='blue')
plt.xticks(np.arange(0,4.1,1))
plt.yticks(np.arange(10,45,5))
plt.grid()
plt.plot(x,y)

x=np.array([8,4,3,2,4,5,6,3,4,2])
y=np.array([4,5,3,5,6,7,8,3,4,3])
plt.subplot(2, 3, 5)
sizes=np.array([10,20,30,40,50,60,70,80,90,20])
color=np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown"])

plt.scatter(x,y,c=color,s=sizes,alpha=0.5)



x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 6)
plt.bar(x,y,color='red',width=0.1)


plt.subplots_adjust(
    left=None, right=None, bottom=None, top=None,
    wspace=0.8, 
    hspace=0.5   
)
plt.suptitle("Subplots")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 20, 20,25])
mycolor=["hotpink","yellow","blue","#880806"]
mylabels = ["Apples", "Bananas", "Cherries","strawberries"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,startangle=90,explode=myexplode,shadow=True,colors=mycolor)
plt.legend()
plt.show() 
```
